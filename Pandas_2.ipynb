**Applying Functions**

```python
#Function of salary category


def sal_cat(sal):
  if sal<6000:
    return "Entry Level"
  elif sal<10000:
    return "Intermediate Level"
  else:
    return "Expert Level"

df['Salary Category'] =df["SALARY"].apply(sal_cat)

df
```

**Data Analysis & Visualization**

***Descriptive Statistics with Pandas***

```python
# Measures of Central Tendency


# MEAN, MEDIAN MODE
mean_MS = df['Monthly Salary'].mean()
median_MS = df['Monthly Salary'].median()
mode_MS = df['Monthly Salary'].mode()


print(f"Mean Salary:{mean_MS}, Median Salary:{median_MS}, Mode Salary:{mode_MS}")
```

```python
#Measure of dispersion

sal_range = df['Monthly Salary']. max() - df['Monthly Salary']. min()

sal_var = df['Monthly Salary'].var()

sal_std = df['Monthly Salary']. std()

print(f"Salary Range: {sal_range}, Variance: {sal_var}, Std Deviation: {sal_std}")
```

**Aggregation Functions**

sum(), mean(), median() ........

```python
# Total Salary by dept

sal_dpt = grp_df['Monthly Salary'].sum()

sal_dpt
```

```python
#Multiple Aggregation Fn

agg_fn ={
    "Monthly Salary":['sum','mean']
}

res= grp_df.agg(agg_fn)

res
```

```python
#Custom Aggregation Function

def sal_range(series):
  return series.max() -series.min()

res_agg = grp_df['Monthly Salary'].agg(sal_range)

res_agg
```

```python
#Grouping Multiple Columns

grp_col = df.groupby(['Department','Job Status'])

grp_col

for key, value in grp_col:
  print(key,value)
```

```python
#Grouping with Time based Data

df['Joining Date'] =pd.to_datetime(df['Joining Date'])

grp_time = df.groupby(pd.Grouper(key = "Joining Date", freq ='M'))

grp_time

for key, value in grp_time:
  print(key,value)
```

```python
# Applying Different Agg Fns to Diff Colums



agg_cols = {
    "Monthly Salary": 'sum',
    "Monthly Salary":'mean'
}

res_col = grp_df.agg(agg_cols)
res_col
```

```python
#Filtering Groups

fil_df = grp_df.filter(lambda x:x['Monthly Salary'].sum()>10000)
fil_df
```
